## Importing all the librabray which we need

```
import cv2
import random
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
from sklearn.model_selection import train_test_split
%matplotlib inline
```

## Generate Test and Training Function

```
DATASET_DIR='dataset/'
```

```
classes=os.listdir(DATASET_DIR)
print(classes)
```

```
data=[]
for cname in classes:
  path=os.path.join(DATASET_DIR,cname)
  imgs=os.listdir(path)
  data.extend([[os.path.join(path,imgname),cname] for imgname in imgs])
  print("{} : {}".format(cname,len(imgs)))

data=np.array(data)
```

```
random.shuffle(data)
X=data[:,0]
y=data[:,1]
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2, 
                                                    random_state=0)
```
